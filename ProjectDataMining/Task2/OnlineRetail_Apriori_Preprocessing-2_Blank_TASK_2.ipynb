# Step 1: Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
OnlineRetail = pd.read_csv("D:\\VS_CODES\\DataMining\\ProjectDataMining\\Dataset\\OnlineRetail.csv", encoding='ISO-8859-1')
OnlineRetail
```

# Step 2: Drop Missing Invoices or Descriptions

```python
OnlineRetail.dropna(subset=['InvoiceNo', 'Description'], inplace=True)
OnlineRetail
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
OnlineRetail = OnlineRetail[~OnlineRetail['InvoiceNo'].astype(str).str.startswith('C')]
OnlineRetail
```

# Step 4: Keep Only Positive Quantities

```python
OnlineRetail = OnlineRetail[OnlineRetail['Quantity'] > 0]
OnlineRetail
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
OnlineRetail['Description'] = OnlineRetail['Description'].str.lower().str.strip()
OnlineRetail
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
OnlineRetail[OnlineRetail["Country"] == "United Kingdom"]
```

# Step 7: Remove Duplicates

```python
OnlineRetail = OnlineRetail.drop_duplicates()
OnlineRetail
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
OnlineRetail['TotalPrice'] = OnlineRetail['Quantity'] * OnlineRetail['UnitPrice']
OnlineRetail
```

# Step 9 : Check for Top 10 Selling Products

```python
top_10_products = OnlineRetail.groupby('Description')["Quantity"].sum().sort_values(ascending=False).head(10)
print(top_10_products)
```

# Step 10 : Identify Most Active Customers

```python
# most_active_customers = (
#     OnlineRetail.groupby('CustomerID')['InvoiceNo']
#     .count()
#     .sort_values(ascending=False)
#     .head(10)
# )
# print(most_active_customers)

most_active_customers = OnlineRetail.groupby("CustomerID")["InvoiceNo"].count().sort_values(ascending=False).head(10)
print(most_active_customers)
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
OnlineRetail['InvoiceDate'] = pd.to_datetime(OnlineRetail['InvoiceDate'])
OnlineRetail
```

```python
# Extract Month and Hour
OnlineRetail['InvoiceMonth'] = OnlineRetail['InvoiceDate'].dt.month
OnlineRetail['InvoiceHour'] = OnlineRetail['InvoiceDate'].dt.hour
OnlineRetail
```

```python
# Popular Months
popular_months = (
    OnlineRetail.groupby('InvoiceMonth')['Quantity']
    .sum()
    .sort_values(ascending=False)
)
print(popular_months)
```

```python
# Popular Purchase Hours
popular_hours = (
    OnlineRetail.groupby('InvoiceHour')['Quantity']
    .sum()
    .sort_values(ascending=False)
)
print(popular_hours)
```

```python
# Pick a small sample of 5 invoices
sample_invoices = OnlineRetail['InvoiceNo'].unique()[:5]  # first 5 unique invoices

# Filter the dataset to include only those invoices
sample_data = OnlineRetail[OnlineRetail['InvoiceNo'].isin(sample_invoices)]

# Preview sample
sample_data[['InvoiceNo', 'Description', 'Quantity']].head()
```

```python
# Create basket format from sample
mini_basket = (
    sample_data
    .groupby(['InvoiceNo', 'Description'])['Quantity']
    .sum()
    .unstack()
    .fillna(0)
    .applymap(lambda x: 1 if x > 0 else 0)
)

# Optional: Rename index to "Transaction" like the image
mini_basket.index.name = 'Transaction'

# Display result
mini_basket.head()
```

# Step:13 : Save Basket Format into CSV file

```python
mini_basket.to_csv("mini_basket.csv")
```
